**How we answer this question:** Find the top 5 countries with the highest average percentage of employment in agriculture and compare it with their employment in industry.

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 
```

### Analyzing to get the answer

Step 01: Construct a dataframe that averages the relevant indicators for employment sectors for each country.

```python
# YOUR CODE HERE
# Create a new dataframe for employment analysis
employment_sector_df = pd.DataFrame()

# Calculate the averages for each relevant feature by country
employment_sector_df["Avg.Employment in Agriculture(%)"] = cleaned_df.groupby("Country")["Employment in Agriculture(%)"].mean()
employment_sector_df["Avg.Employment in Industry(%)"] = cleaned_df.groupby("Country")["Employment in Industry(%)"].mean()

# Set the index as a column - countries
employment_sector_df["Country"] = employment_sector_df.index

# Drop the index
employment_sector_df.reset_index(drop=True, inplace=True)

# Sort the dataframe based on the average employment in agriculture, descending
employment_sector_df.sort_values("Avg.Employment in Agriculture(%)", inplace=True, ascending=False)

# Take the top 5 countries for visualization
top_employment_sector_df = employment_sector_df.head(5)

# Display
top_employment_sector_df


```

Step 02: Visualize the data using a bar chart to compare these top countries.

```python
# Choose a country for the demonstration, e.g., 'CountryName'
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(12, 6))

# Plot the average employment in agriculture
sns.barplot(x="Avg.Employment in Agriculture(%)", y="Country", data=top_employment_sector_df,
            label="Agriculture", color="g")

# Plot the average employment in industry on top of the agriculture
sns.barplot(x="Avg.Employment in Industry(%)", y="Country", data=top_employment_sector_df,
            label="Industry", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 100), ylabel="",
       xlabel="Average Employment in Agriculture vs Industry (%)")
sns.despine(left=True, bottom=True)

# Add a title to the plot
plt.title('Top 5 Countries with Highest Average Employment in Agriculture vs Industry', size=16)

# Show the plot
plt.show()
```

**MY OBSERVATION**

- **Agricultural Dominance:** In all five countries listed, the agriculture sector employs a larger percentage of the population than the industrial sector. This indicates a significant reliance on agriculture as the backbone of their economies.
- **Highest in Agriculture:** `Lao PDR` appears to have the highest percentage of its employment in agriculture, surpassing 80% of the workforce, with a very small portion employed in industry.
- **Significant Agricultural Employment:** Nepal and Bhutan also have a high percentage of their populations employed in agriculture, both showing slightly more than 60% agricultural employment.
- **Comparatively Lower in Industry:** `Cambodia` and `Afghanistan` show a higher percentage of employment in the industry compared to the other three countries, but agriculture remains the dominant employment sector.
- **Economic Structure Implications:** The chart indicates that these top 5 countries have an economy that is strongly dependent on agriculture, with industrial sectors being much smaller in comparison.

In conclusion, the top 5 countries with the highest average percentage of employment in agriculture are `Lao PDR, Nepal, Bhutan, Cambodia, and Afghanistan`. In these countries, the employment percentage in agriculture is substantially higher than in industry, emphasizing a pronounced reliance on agriculture for livelihoods. This economic structure could suggest that these countries are potentially vulnerable to agricultural volatility due to factors such as climate change, market prices, and international trade policies. However, it could also indicate a strong capacity for sustainable agricultural practices and potential for growth in agritourism and organic markets. The lower relative percentage of industrial employment also suggests that these economies might be less industrialized and could either represent an untapped potential for industrial development or a strategic focus on sustainable agricultural development.

**A Possible Question:** What is the distribution of renewable energy consumption compared to fossil fuel consumption across these countries?

**Why This Question Is Relevant:** Understanding the distribution of renewable versus fossil fuel energy consumption can offer insights into the sustainability practices of different countries and indicate how prepared they are for a transition to cleaner energy sources. This information is crucial in the context of global efforts to combat climate change and reduce greenhouse gas emissions.

**Answering this question will:**
- Highlight countries that are leading in renewable energy consumption, setting an example for sustainable energy use.
- Indicate countries with high reliance on fossil fuels, possibly identifying areas where policy intervention could encourage more renewable energy usage.
- Show global trends in energy consumption which are critical for assessing progress towards international climate goals.

**How we answer this question:**
- Extract data on renewable energy consumption percentages and fossil fuel consumption percentages for each country.
- Create a composite bar chart where each country is represented by two bars side by side, one for renewable and one for fossil fuel consumption, to directly compare the two within each country.
- Alternatively, create a bar chart for select countries to represent the proportion of renewable to fossil fuel consumption, which can quickly communicate the share of each energy type in a country's energy profile.

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 
```

### Analyzing to get the answer

Step 01: Construct a dataframe that aggregates the average female and male employment percentages in agriculture and industry for each country.

```python
# YOUR CODE HERE
# Create a new dataframe for average renewable and fossil fuel consumption
energy_consumption_df = pd.DataFrame()

# Calculate average renewable and fossil fuel consumption for each country
energy_consumption_df["Avg.Renewable Consumption (%)"] = cleaned_df.groupby("Country")["Renewable Energy Consumption (%)"].mean()
energy_consumption_df["Avg.Fossil Fuel Consumption (%)"] = cleaned_df.groupby("Country")["Fossil Fuel Consumption (%)"].mean()

# Set the index as a column - countries
energy_consumption_df["Country"] = energy_consumption_df.index

# Drop the index
energy_consumption_df.reset_index(drop=True, inplace=True)

# Sort the dataframe based on renewable energy consumption
energy_consumption_df.sort_values("Avg.Renewable Consumption (%)", inplace=True, ascending=False)

# Display the top 5 countries
energy_consumption_df.head()
```

Step 02: Visualize the data using bar charts.


```python
# Now we melt the dataframe to have a suitable structure for sns.barplot
# Make sure to get the top 10 countries sorted by renewable consumption
top_countries = energy_consumption_df.head(5)
melted_df = top_countries.melt(id_vars="Country", value_vars=['Avg.Renewable Consumption (%)', 'Avg.Fossil Fuel Consumption (%)'], var_name="Type", value_name="Average Consumption (%)")

# Plotting the distribution for top countries
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
ax = sns.barplot(x='Country', y='Average Consumption (%)', hue='Type', data=melted_df)

plt.xticks(rotation=45)
plt.title("Top 5 Countries by Renewable and Fossil Fuel Energy Consumption")
plt.ylabel("Average Consumption (%)")
plt.xlabel("Country")
plt.legend(title='Energy Type')
plt.tight_layout() # This will adjust the plot to ensure everything fits without overlap
plt.show()
```

**MY OBSERVATION**

- **Renewable vs. Fossil Fuel Energy Consumption:** For each country shown, the consumption of renewable energy is higher than that of fossil fuels. This suggests a strong orientation towards renewable energy sources in these countries.
- **High Renewable Energy Utilization:** `Bhutan` and `Nepal` stand out with the highest renewable energy consumption percentages compared to fossil fuels, indicating that these countries have leveraged their natural resources.
- **Moderate Renewable Energy Utilization:** `Myanmar` and `Cambodia` show a good balance between renewable and fossil fuel consumption but still lean significantly towards renewable sources.
- **Lower Gap in Lao PDR:** `Lao PDR` exhibits a lower gap between renewable and fossil fuel energy consumption. This might suggest that while renewable energy is predominant, there's still a considerable reliance on fossil fuels compared to the other countries displayed.
- **Uniformity in Preference for Renewable Sources:** Across all these countries, there is a consistent pattern where renewable energy consumption is favored over fossil fuel consumption. This could reflect geographic, economic, and policy factors that encourage the use of renewable resources.

In conclusion, the distribution of renewable energy consumption compared to fossil fuel consumption across these countries leans significantly towards renewable energy. It indicates that these countries are potentially well-positioned to transition to cleaner energy sources, which is crucial for reducing greenhouse gas emissions and combating climate change. This distribution can be seen as a positive sign of sustainable practices and may reflect policy decisions, availability of renewable resources, and investment in renewable energy infrastructure. Each country's specific context would influence their respective energy profiles, but overall, the focus on renewable is apparent.
